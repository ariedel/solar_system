```python
import glob
import numpy as np
from matplotlib import pyplot as plt
```

```python
filenames = glob.glob('data/inflammation-*.csv')

averages = []

for filename in filenames:
    datum = np.loadtxt(fname=filename, delimiter=',')
    fig = plt.figure(figsize=(17,6))
    ax1 = fig.add_subplot(141)
    ax2 = fig.add_subplot(142)
    ax3 = fig.add_subplot(143)
    ax4 = fig.add_subplot(144)

    ax1.plot(np.mean(datum,axis=0))
    ax2.plot(np.max(datum,axis=0))
    ax3.plot(np.min(datum,axis=0))
    ax4.imshow(datum,cmap='inferno')
    averages.append(np.mean(datum,axis=0))

    ax1.set_ylabel('average')
    ax2.set_ylabel('minimum')
    ax3.set_ylabel('maximum')
    ax4.set_title('{0:}'.format(filename))
    plt.tight_layout()
    plt.show()
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(averages[0]-averages[1])
ax.set_title(r'$\Delta$ average')
plt.show()
```

```python
filenames = glob.glob('data/inflammation-*.csv')

values = []

for filename in filenames:
    datum = np.loadtxt(fname=filename, delimiter=',')
    values.append(datum)
    if (np.max(datum,axis=0)[0] == 0) and (np.max(datum,axis=0)[20] == 20):
        print('Suspicious looking maxima!')
    elif np.sum(np.min(datum,axis=0)) == 0:
        print('Minima add up to zero!')
    else:
        print('Reasonable data')

average = np.mean(values,axis=1)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(average)
ax.set_title(r'Mean per patient')
plt.show()
```

```python
def analyze(filename):
    '''plots average, max, min, and 2D image representation of a given dataset'''
    datum = np.loadtxt(fname=filename, delimiter=',')
    fig = plt.figure(figsize=(17,5))
    ax1 = fig.add_subplot(141)
    ax2 = fig.add_subplot(142)
    ax3 = fig.add_subplot(143)
    ax4 = fig.add_subplot(144)

    ax1.plot(np.mean(datum,axis=0))
    ax2.plot(np.max(datum,axis=0))
    ax3.plot(np.min(datum,axis=0))
    ax4.imshow(datum,cmap='inferno')
    averages.append(np.mean(datum,axis=0))

    ax1.set_ylabel('average')
    ax2.set_ylabel('minimum')
    ax3.set_ylabel('maximum')
    ax4.set_title('{0:}'.format(filename))
    plt.tight_layout()
    plt.show()
```

```python
analyze('data/inflammation-02.csv')
```

```python
def fence(original,wrapper):
    '''Returns the original string surrounded by the fence string'''
    return('{1:}{0:}{1:}'.format(original,wrapper))
```

```python
print(fence('M',"**"))
```

```python
def outer(instring):
    '''Returns the first and last character of the string'''
    return("{0:}{1:}".format(instring[0],instring[-1]))
```

```python
print(outer('M'))
```

```python
def rescale(array):
    '''Rescales an array of values to be between 0 and 1'''
    minval = np.min(array)
    maxval = np.max(array)
    return (array-minval)/(maxval-minval)
```

```python
array = [0, 16, 24, 32, 64, 128, 22, 9, 83, 256, 144,-81, -42]
```

```python
print(rescale(array))
```

```python
    def _sersic_func(y, x, major, minor, index):
        """
        Implement Sersic intensity profile in a way that scipy.integrate.dblquad can use since it is passed there
        as a callable. We need to integrate this function to normalize it properly for flux outside the calculation FOV.

        Parameters
        ----------
        y: float or numpy.ndarray
            Y values for evaluating function
        x: float or numpy.ndarray
            X values for evaluating function
        major: float
            Major axis scale length
        minor: float
            Minor axis scale length
        index: float
            Sersic index

        Returns
        -------
        profile: float or numpy.ndarray
            Float or array containing evaluated Sersic profile
        """
        dist = np.sqrt((x / major)**2.0 + (y / minor)**2.0)
        profile = np.exp(-dist**(1.0 / index))
        return profile

```
